```python
import os
import librosa
from matplotlib import pyplot as plt
```

(.Wav)Transform to Waveform 


```python
filelist = []
data_dir = r'D:\Data\Chen'
for  filename in os.listdir(data_dir):
    filelist.append(filename)
# print(filelist)
class_dir = os.path.join(data_dir, filelist[0])
# print(class_dir)
```

```python
y ,sr = librosa.load(class_dir)
```

```python
plt.plot(y)
plt.title('Signal')
plt.xlabel;('samples')
plt.ylabel('Amplitude')
```

Fast Fourier Transform


```python
import numpy as np 
n_fft=2048
ft = np.abs(librosa.stft(y[:n_fft], hop_length=n_fft+1))

plt.plot(ft)
plt.title('Spectrum')
plt.xlabel('Frequency Bin')
plt.ylabel('Amplitude')
```

Spectrogram

```python
import librosa.display

spec = np.abs(librosa.stft(y, hop_length=512))
spec = librosa.amplitude_to_db(spec, ref=np.max) #using log to compute the amplitude to db

librosa.display.specshow(spec, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
```

Mel Spectrogram

```python
mel_spect = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=2048, hop_length=1024)
mel_spect = librosa.power_to_db(mel_spect, ref=np.max)
librosa.display.specshow(mel_spect, y_axis='mel', fmax=8000, x_axis='time');
plt.title('Mel Spectrogram');
plt.colorbar(format='%+2.0f dB');
```
